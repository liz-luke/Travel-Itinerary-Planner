```python
df_reviews = df_reviews[(df_reviews['state'] == 'kerala')|(df_reviews['state'] == 'rajasthan')|(df_reviews['state'] == 'uttar pradesh')]
df_reviews.head()
```

```python
def get_sentiment(sentence):
    analysis = TextBlob(sentence)
    
    if analysis.sentiment.polarity > 0:
        return "Positive"
    elif analysis.sentiment.polarity < 0:
        return "Negative"
    else:
        return "Neutral"


df_reviews['Sentiment'] = df_reviews['Review'].apply(get_sentiment)

df_reviews.head()
```

```python
df_reviews['Sentiment'].value_counts()
```

```python
df_rev_rating_mean = pd.DataFrame(df_reviews.groupby('Place')['Rating'].mean())
df_rev_rating_mean.reset_index(inplace = True)
```

```python
df_rev_rating_mean
```

```python
df_sentiment = pd.DataFrame(df_reviews.groupby('Place')['Sentiment'].value_counts())
df_sentiment.head()
```

```python
df_sentiment.rename(columns = {'Sentiment':'count'}, inplace = True)
df_sentiment.reset_index(inplace = True)
```

```python
place_tot_dict = {}
for place in df_sentiment['Place'].unique():
    place_tot_dict[place] = df_sentiment[df_sentiment['Place'] == place]['count'].sum()
place_tot_dict
```

```python
df_sentiment
```

```python
for index, row in df_sentiment[df_sentiment['Sentiment'] == 'Positive'].iterrows():
    positive_count = row['count']
    place_tot_dict[row['Place']] = positive_count/place_tot_dict[row['Place']]
place_tot_dict
```

```python
data = {'place': list(place_tot_dict.keys()), 'ratio': list(place_tot_dict.values())}
ratio_df = pd.DataFrame(data)
ratio_df
```

```python
combined_df = df_rev_rating_mean.merge(ratio_df, left_on = 'Place', right_on = 'place')
combined_df.drop(['place'], inplace = True, axis = 1)
combined_df
```
